```python
import requests
import pandas
import re
from bs4 import BeautifulSoup
```

### Sending a Request to Scraping the Page 

```python
r = requests.get('https://docs.python.org/3/library/')
result = r.text
```

```python
soup = BeautifulSoup(result,'lxml')
```

### Result of the Request in the form of Html

```python
soup
```

### Using Html Attribute to Scrape the list of the Python Standard functions it can also be done using Tags

```python
links = soup.body.findAll(attrs={"class" : "reference internal"})
links
```

### Removing first item as it was not Relavent 

```python
links.pop(0)
```

### Extracting the text to List

```python
final_result = []
for item in links:
    item = item.text
    final_result.append(item)
final_result
```
